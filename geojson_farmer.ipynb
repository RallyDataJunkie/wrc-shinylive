```python
url="https://webappsdata.wrc.com/srv/wrc/json/api/wrcsrv/queryMeta?t=%22Event%22&p=%7B%22n%22%3A%22category%22%2C%22v%22%3A%22WRC%22%7D&maxdepth=1"

import requests
feed = requests.get(url).json()
geo_paths = []
for f in feed:
    for m in f["_meta"]:
        if m["n"]=="kmlfile":
            geo_paths.append(m["v"])
geo_paths
```

```python
import kml2geojson
import re
from io import StringIO
import os
import json


WRC_ASSETS_PATH = "https://webapps2.wrc.com/2020/web"
WRC_KML_PATH = WRC_ASSETS_PATH + "/live/kml/{kmlfile}.xml"

geodir = "geodir"
if not os.path.exists(geodir):
    os.makedirs(geodir)

for geo_path in geo_paths:
    kmlurl = WRC_KML_PATH.format(kmlfile=geo_path)
    text = StringIO(
            re.sub(
                r"[\u200b\u200e\u200f]",
                "",
                requests.get(kmlurl).content.decode("utf-8"),
            )
        )
    with open(f"{geodir}/{geo_path}.json", "w") as f:
        json.dump(kml2geojson.main.convert(text), f)
```

```python
from pathlib import Path
from zipfile import ZipFile, ZIP_DEFLATED

# Path to your directory containing GeoJSON files
geojson_dir = Path("./resources")

# Loop over all .geojson files in the directory
for geojson_file in geojson_dir.glob("*.json"):
    zip_name = geojson_file.with_suffix(".zip")  # Replace .geojson with .zip
    with ZipFile(zip_name, "w", compression=ZIP_DEFLATED) as zipf:
        zipf.write(
            geojson_file, arcname=geojson_file.name
        )  # store with just the filename
    print(f"Created {zip_name}")
```
