```python
import time

%run wrc_rallydj/livetiming_api2.py

for year in [2025]:
    wrc = WRCTimingResultsAPIClientV2(year=year,newDB=True, dbname=f"wrc{year}seasonarchive.db")

    championships = wrc.getChampionships().championshipId.tolist()

    for champ in championships:
        wrc.setChampionship(champ)

    season_rounds = wrc.getSeasonRounds()

    for eventname in season_rounds.name.to_list():
        wrc.setEventByName(eventname)
        wrc.eventId, wrc.eventName, wrc.year, wrc.isRallyLive(), wrc.isRallyInDate()
        wrc.getEventShakeDownTimes(wrc.eventId, updateDB=True)
        wrc.getStageTimes(completed=True, raw=False, updateDB=True)
        time.sleep(0.5)
        wrc.getStageOverallResults(completed=True, raw=False, updateDB=True)
        time.sleep(0.5)
        wrc.getPenalties(raw=False, updateDB=True)
        wrc.getRetirements(raw=False, updateDB=True)
        wrc.getStageWinners()

        stageinfo = wrc.getStageInfo()
        for stage in stageinfo.code.to_list():
            time.sleep(0.3)
            wrc.setStageByCode(stage)
            wrc.getSplitTimes()
```
