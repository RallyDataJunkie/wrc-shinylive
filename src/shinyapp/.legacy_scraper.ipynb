# Legacy Scraper

Attempt to archive data before the old API disappears.

```python
from wrc_rallydj.livetiming_api import WRCLiveTimingAPIClient
from requests_cache import NEVER_EXPIRE

wrc = WRCLiveTimingAPIClient(
    use_cache=True, backend="sqlite", expire_after=NEVER_EXPIRE, cache_name="backup2024"
)
```

```python
wrc.initialise(year=2024)
```

```python
wrc.full_calendar
```

```python
wrc.seasonId
```

```python
for c in  ["wrc",  "wrc2", "wrc3", "jwrc", "wrc2c", "mcup"]:
    for t in ["driver", "codriver", "team", "manufacturer"]:
        wrc.getChampionship(championship_type=t, championship=c)
```

```python
for c in  ["wrc",  "wrc2", "wrc3", "jwrc", "wrc2c", "mcup"]:
    wrc.getResultsCalendar(year=2024, seasonId=28,championship=c)
```

```python
for r in wrc.full_calendar.iterrows():
    print(r[1]["eventId"], r[1]["rallyId"])
    wrc.getItinerary(eventId=r[1]["eventId"])
    wrc.getStartlist(eventId=r[1]["eventId"])

    wrc.getPenalties(eventId=r[1]["eventId"])

    wrc.getRetirements(eventId=r[1]["eventId"])

    s = wrc.getStageDetails(eventId=r[1]["eventId"], rallyId=r[1]["rallyId"])

    for c in ["wrc", "wrc2", "wrc3", "jwrc", "wrc2c", "mcup"]:
        wrc.getStageWinners(eventId=r[1]["eventId"], championship=c)
        for _s in s.iterrows():
            wrc.getOverall(
                eventId=r[1]["eventId"],
                rallyId=r[1]["rallyId"],
                stageId=_s[1]["stageId"],
                championship=c,
            )
            try:
                wrc.getStageTimes(
                eventId=r[1]["eventId"],
                rallyId=r[1]["rallyId"],
                stageId=_s[1]["stageId"],
                championship=c,)
            except:
                print("error getStageTimes")
            try:
                wrc.getSplitTimes(
                    eventId=r[1]["eventId"],
                    rallyId=r[1]["rallyId"],
                    stageId=_s[1]["stageId"],
                    championship=c,
                )
            except:
                print("error getSplitTimes")
```
